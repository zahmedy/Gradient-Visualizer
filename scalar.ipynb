```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (3*x - 1)**2

def df_dx(x):
    return 18*x - 6
    raise NotImplementedError
```

```python
xs = np.linspace(-2, 2, 400)
print(xs)
ys = f(xs)
print(ys)
dys = df_dx(xs)
print(dys)
```

```python
plt.figure()
plt.plot(xs, ys, label="f(x)")
plt.plot(xs, dys, label="f'(x)")

x0 = 1/3
y0 = f(x0)
slope = df_dx(x0)
print(f"slope: {slope}")
tangent = y0 + slope * (xs - x0)

plt.plot(xs, tangent, '--', label=f"Tangent at x0={x0}")
plt.scatter([x0], [y0], color='red')  # the point of tangency
plt.legend(); plt.title("f and f")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.plot(xs, ys, label='f(x)')

# pick few x position to visualize slope 
x_points = [-0.5, 0.33, 1]
colors = ['red', 'green', 'blue', 'black']

for x0, c in zip(x_points, colors):
    y0 = f(x0)
    slope = df_dx(y0)
    tangent = y0 + slope * (xs - x0)
    plt.plot(xs, tangent, '--', color=c, label=f"x0={x0:.2f}, slope={slope:.1f}")
    plt.scatter([x0], [y0], color=c)

plt.legend()
plt.title("Tangent lines showing slope direction and sign")
plt.show()
```

```python
# gradient descent parameters
x = 2.0         # start point
lr = 0.05       # learning rate
steps = 20

xs_path = [x]
for i in range(steps):
    grad = df_dx(x)
    x = x - lr * grad
    xs_path.append(x)

# visualize
# xs = np.linspace(-1, 2, 400)
# ys = f(xs)

plt.figure(figsize=(7,5))
plt.plot(xs, ys, label='f(x)')
plt.scatter(xs_path, f(np.array(xs_path)), color='red', zorder=5)
plt.plot(xs_path, f(np.array(xs_path)), '--', color='red', label='Descent path')
plt.legend()
plt.title("Gradient Descent Path on f(x) = (3x - 1)^2")
plt.show()
```
