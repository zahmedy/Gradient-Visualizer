```python
import numpy as np
import matplotlib.pyplot as plt

# vectors 
a = np.array([3,1])
b = np.array([2,-2])

print(f"vector a: {a} \n vector b: {b}")

# dot product and projection 
dot = np.dot(a, b)
b_norm = np.linalg.norm(b)
proj_length = dot / b_norm
proj_vector = (proj_length / b_norm) * b

# plot set up
plt.figure(figsize=(6, 6))
plt.axhline(0, color='grey', lw=1)
plt.axvline(0, color='grey', lw=1)

# draw vectors
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='red', label='a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='blue', label='b')
plt.quiver(0, 0, proj_vector[0], proj_vector[1], angles='xy', scale_units='xy', scale=1, color='green', label='projection of a on b')

plt.xlim(-1, 4)
plt.ylim(-3,3)
plt.legend()
plt.gca().set_aspect('equal')
plt.title(f"Dot product = {dot:.1f}, projection length = {proj_length:.2f}")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# define function and gradient
def f(x, y):
    return x**2 + y**2

def grad_f(x, y):
    return np.array([2*x, 2*y])

# grid for visualization
x_vals = np.linspace(-3, 3, 20)
y_vals = np.linspace(-3, 3, 20)
X, Y = np.meshgrid(x_vals, y_vals)
U, V = 2*X, 2*Y  # gradient components

plt.figure(figsize=(6,6))
plt.contour(X, Y, f(X,Y), levels=10, cmap='coolwarm')
plt.quiver(X, Y, U, V, color='gray', alpha=0.5)
plt.title("Gradient field for f(x, y) = x + y")
plt.xlabel("x"); plt.ylabel("y")

# gradient descent simulation
lr = 0.2
steps = 10
point = np.array([2.5, 2.0])
path = [point.copy()]

for _ in range(steps):
    g = grad_f(point[0], point[1])
    point = point - lr * g
    path.append(point.copy())

path = np.array(path)
plt.plot(path[:,0], path[:,1], 'ro--', label='Descent path')
plt.legend()
plt.gca().set_aspect('equal')
plt.show()
```
